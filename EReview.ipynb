```python
def conv_fname(x):
  if len(x) < 11:
    x = x[0] + x[3:9] + '001' + 'p'
  else:
    x = x[0] + x[3:9] + x[-3:] + 'p'

  return x.upper()
```

```python
dcl['filename'] = dcl['docno'].apply(conv_fname)
```

```python
dcl['filename'] = dcl['filename'] + dcl['rev'] + '.pdcl'
```

```python
dcl['filename'] = dcl['filename'].str.upper()
```
